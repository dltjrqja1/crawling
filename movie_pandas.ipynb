```python
import pandas as pd
```

```python
movie = pd.read_csv('./movie_df.csv',sep=',',index_col=[0])
```

```python
movie
```

```python
movie = movie.fillna(0)
```

```python
movie
```

```python
total_movie = movie
```

```python
movie = movie[(movie['Advance_rate']>0.01)]
movie
```

```python
movie.groupby('age').describe()
```

```python
movie.sort_values(by='rating',axis=0,ascending=False)
```

```python
movie_split = movie.genres.str.split(' ',expand=True)
```

```python
movie_split.columns = ['genre','genre','genre','genre','genre','genre']
movie_long = pd.concat([movie, movie_split],axis=1)
```

```python
movie_long = movie_long.drop(['genres'],axis=1)
movie_long
```

```python
movie_long = movie_long.fillna(' ')
```

```python
movie_long
```

```python
movie_long = pd.melt(movie_long, id_vars=['title','age','Advance_rate','rating'],var_name='tmp',value_name='genre')
```

```python
movie_long
```

```python
movie_long = movie_long.drop(['tmp'],axis=1)
```

```python
movie_long = movie_long[movie_long.genre!=" "]
movie_long
```

```python
movie_long.groupby('genre').describe()
```
